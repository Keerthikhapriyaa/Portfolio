```python
# Step 2: College Student Assistant
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.tools import Tool
from langchain.agents import initialize_agent, AgentType

# Set your Google API key
os.environ["GOOGLE_API_KEY"] = "AIzaSyDLzBYSEFHyRWGPTTEyfcBbg7xmD7aV0lE"

# Initialize LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)

# Define tools
exam_tool = Tool(
    name="Exam Schedule Tool",
    func=lambda x: """
Upcoming Exams:
1. Data Structures – March 10
2. DBMS – March 14
3. Operating Systems – March 18
""",
    description="Use this tool to get exam schedule"
)

marks_tool = Tool(
    name="Internal Marks Tool",
    func=lambda x: """
Internal Marks:
DS – 25/30
DBMS – 27/30
OS – 22/30
""",
    description="Use this tool to get student internal marks"
)

attendance_tool = Tool(
    name="Attendance Tool",
    func=lambda x: """
Attendance Percentage:
Data Structures – 88%
DBMS – 92%
Operating Systems – 85%
Overall Attendance – 88%
""",
    description="Use this tool to get student attendance details"
)

# Initialize agent
agent = initialize_agent(
    tools=[exam_tool, marks_tool, attendance_tool],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# Run the agent
response = agent.run("Show my exam schedule, internal marks and attendance")
print(response)
```
